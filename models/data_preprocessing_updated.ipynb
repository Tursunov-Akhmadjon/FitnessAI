```python
df.dropna(subset=['VO2 Max', 'Body Fat (%)', 'Water Intake (liters)'], inplace=True)
```

```python
num_cols = df.select_dtypenum_cols = df.select_dtypes(include='number').columns.drop(['Workout Type', 'Calories Burned'], errors='ignore')
cat_cols = df.select_dtypes(include='object').columns
```

```python
num_features = Pipeline(steps=[
  ('imputer', SimpleImputer(strategy='mean')),
  ('scaler', StandardScaler())
])

cat_features = Pipeline(steps=[
  ('imputer', SimpleImputer(strategy='most_frequent')),
  ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(transformers=[
  ('num', num_features, num_cols),
  ('cat', cat_features, cat_cols)
])

pipeline = Pipeline(steps=[
  ('preprocessor', preprocessor)
])
```
